# converting from  Unix Nano format to Datetime

```python
referral= referral.astype({'created_at':'datetime64[ns]', 'updated_at': 'datetime64[ns]'})
```

```python
referral
```

```python
referral.info()
```

```python
referral["created_at"] = pd.to_datetime(referral["created_at"])
```

```python
referral["updated_at"] = pd.to_datetime(referral["updated_at"])
```

```python
referral["updated_at"]
```

```python
referral['created_date']= pd.DatetimeIndex(referral.created_at).date
referral['updated_date']= pd.DatetimeIndex(referral.updated_at).date
```

```python
referral= referral.drop(['created_at','updated_at'],axis=1)
```
